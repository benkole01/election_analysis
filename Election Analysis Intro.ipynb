The data for the polls is obtained from HuffPost Pollster. It has some political datasets to play with.

We will use requests module to import some data from the web.

We will also use StringIO to work with csv data we get from HuffPost. StringIO provides a convenient means of working with text in memory using the file API.

```python
# for grabbing data from the web(HTTP capabilities)
import requests

# StringIO to work with the csv file, the DataFrame will require a .read() method
from io import StringIO
```

```python
# this is the url link for the poll data in csv form
url = 'http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv'

# use requests to get the information in text form
source = requests.get(url).text

# use StringIO to avoid an IO error with pandas
poll_data = StringIO(source)
```

Now that we have our data,we can set it as a DataFrame

```python
# set poll data as pandas DataFrame
poll_df = pd.read_csv(poll_data)
```
